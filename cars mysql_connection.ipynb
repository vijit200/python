# Import lib

```python
import pandas as pd
import mysql.connector as conn
from mysql.connector import Error
import csv
import logging as lg
```

# Converting Database Into DataFrame

```python
lg.basicConfig(filename = "Car.log", level = lg.INFO, format = "%(asctime)s %(levelname)s %(message)s")
car = pd.read_csv("car.csv",index_col=False,delimiter=",")
print(car.head())
lg.info("data frame created")
lg.info("{}".format(car.head()))
```

# Creating Connection & data base 

```python
try :
    mydb = conn.connect(host = "localhost",user = "root", passwd = "Vijit699@")
    if mydb.is_connected():
        cur = mydb.cursor()
        cur.execute("CREATE DATABASE IF NOT EXISTS Cars")
        print("database created...")
        lg.info("Server connection done")
        lg.info("database is created")
except Exception as e:
    lg.error(str(e))
```

# Creating Table & Inserting

```python
cur.execute("CREATE TABLE IF NOT EXISTS Cars.CarsData(Buying VARCHAR(25),Price VARCHAR(25),Doors VARCHAR(25),Person VARCHAR(25),Log_boot VARCHAR(25),Safety VARCHAR(25),Class_value VARCHAR(25))")
lg.info("Table is created")
```

```python
#
try :
    for i,row in car.iterrows():
        sql = "INSERT INTO Cars.CarsData VALUES(%s,%s,%s,%s,%s,%s,%s)"
        cur.execute(sql,tuple(row))
        lg.info("inserting values in table")
        mydb.commit()
        lg.info("Values inserted sucessfully")
    print("inserted")
except Exception as e:
    lg.error(e)
```

# Checking Data inserted or not

```python
try:
    if mydb.is_connected():
        lg.info("checking data is inserted")
        cur.execute("SELECT * FROM Cars.Carsdata")
        print(cur.fetchall())
    lg.info("data checked successfully")
        
except Exception as e:
    lg.error(e)
```

# Group By

```python
try:
    cur.execute("SELECT COUNT(*) ,Buying FROM Cars.Carsdata GROUP BY Buying")
    lg.info("grouping data")
    print(cur.fetchall())
except Exception as e:
    lg.error(e)
```

# Filtering Records

```python
try:
    cur.execute("SELECT * FROM Cars.Carsdata WHERE Doors = '4'")
    lg.info("filtering data")
    for i in cur.fetchall():
        print(i)
except Exception as e:
    lg.error(e)
```

# Update

```python
try:
    cur.execute("UPDATE Cars.Carsdata SET Doors = '8' WHERE Doors = '2'")
    lg.info("updating data")
    cur.execute("SELECT * FROM Cars.Carsdata WHERE Doors ='8'")
    for i in cur.fetchall():
        print(i)
except Exception as e:
    lg.error(e)
```

# Delete Table

```python
try:
    lg.info("droping table")
    cur.execute("DROP TABLE IF EXISTS Cars.Carsdata")
    print("table droped..")
    lg.info("table drop succesfully")
    cur.execute("SHOW TABLE Cars.Carsdata")
except Exception as e:
    lg.error(e)
```

# Drop Database

```python
try:
    lg.info("droping database")
    cur.execute("DROP DATABASE IF EXISTS Cars")
    print("Database droped..")
    lg.info("database drop succesfully")
    cur.execute("SHOW DATABASE Cars")
except Exception as e:
    lg.error(e)
```
